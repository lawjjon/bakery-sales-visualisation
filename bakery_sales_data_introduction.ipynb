The Bakery dataset contains transactional records for a small bakery in Korea, and is available on [Kaggle](https://www.kaggle.com/datasets/hosubjeong/bakery-sales?select=Bakery+Sales.csv). The variables are introduced below:

| Column name | Description |
| :---------- | :---------- |
| datetime    | The time and date of order|
| day of week | Day of the week of order (usually closed on Tuesdays)|
| total       | The total order amount (KRW) |
| place       | The customer's location |
|*remaining columns* | The name of the food/drink item |
